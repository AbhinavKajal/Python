### DataFrame And Indexing :-

```python
import pandas as pd
import numpy as np
name1 = pd.Series([1,2,3,4],index=['Abhi','Tannu','king','singh'])
name1 # pd.Series is used for creating Uni-Dimensional Data ( Single Column )
```

```python
name1['king']
```

```python
name2 = pd.Series([10,2,3,14],index = ['Abhi','sam','dean','Tannu'])
name2
```

```python
name1 + name2 # NaN applies for Not A Number
```

```python
_score = pd.DataFrame(np.random.rand(8,5),index='A B C D E F G H'.split(),columns='Score1 Score2 Score3 Score4 Score5'.split())
_score# pd.DataFrame is used to make MultiDimensinal Data 
```

```python
_score['Score1'] 
```

```python
_score[['Score1','Score2','Score3']]
```

```python
_score['Score6'] = _score['Score1']+ _score['Score2']
_score
```
